Importing various libraries

```
#for the web crawling
import requests
import urllib.request
import time
import bs4
from bs4 import BeautifulSoup

#for extracting the text
import re

#for writing to a csv, txt file
import csv
import ast


#for the neural networks
import tensorflow as tf
import numpy as np
import os

#for tagging
import nltk
from nltk import pos_tag
from nltk import RegexpParser
```

Do the initial extraction from the urls and output them to a csv  'ingredients.csv'

```
#make an empty csv
with open('ingredients.csv','w',newline='') as file:
    writer=csv.writer(file)

#cycle through list of urls
for link in urls:

  #crawl and get it
  response = requests.get(link)
  soup = BeautifulSoup(response.text, 'html.parser')

  IngrWithSectsList = soup.p.parent.contents
  seenP = 0
 
  
  #mine the text
  ingredientList = []

  for line in IngrWithSectsList:#We're going to ignore the not-tags. We are going to keep ingredients, until we get to a header *after* an ingredient, at which point we skip to the next recipe
    if isinstance(line,bs4.element.Tag):##If it's a tag


      lineClass = line['class'][0]#Class is going to tell us what type of tag it is. Without the 0 it still registers as a list
      if seenP ==1 and lineClass.find('Headline') != -1: #if we've already seen an ingredient and now we're at a new section-> break
        break
    
    #If we made it through; that is, we either haven't seen a p tag yet or we're not at a section head
      if lineClass == 'o-Ingredients__a-Ingredient': #if it's an ingredient
        seenP = 1
        ingredientList.append(line.text.lower())#including lower here, for regex reasons later on


  #write to the csv
  with open('ingredients.csv','a',newline='') as file:
    writer=csv.writer(file)
    writer.writerow(ingredientList)
```

```
import html

with open("ingredients.csv", "r") as infile, open("interFixedIngredients.csv", "w") as outfile:
    reader = csv.reader(infile)
    writer = csv.writer(outfile)
    forRemoval = set('?')
    #forConversion = 
    for row in reader:
        newrow = [''.join('\'' if c in forRemoval else c for c in entry) for entry in row]
        writer.writerow(newrow)


# open your csv and read as a text string
with open('interFixedIngredients.csv', 'r') as f:
    my_csv_text = f.read()

find_str = '&amp;'
replace_str = '&'

# substitute
new_csv_str = re.sub(find_str, replace_str, my_csv_text)

# open new file and save
new_csv_path = 'fixedIngredients.csv' # or whatever path and name you want
with open(new_csv_path, 'w') as f:
    f.write(new_csv_str)
```

```
card_str_to_int = {'one': '1', 'two': '2', 'three': '3', 'four': '4', 'five': '5',
                   'six': '6', 'seven': '7', 'eight': '8', 'nine': '9', 'ten': '10'}

cup_conv = {'stick': 2, 'sticks': 2, #'cup': 1, 'cups': 1,
            #'tablespoon': 16, 'tablespoons': 16,
            #'teaspoon': 48, 'teaspoons': 48,
            'ounce': 8, 'ounces': 8, 'pound': .5, 'pounds': .5}

cup_conv_p = {'stick': 2, 'sticks': 2, 'cup': 1, 'cups': 1,
            'tablespoon': 16, 'tablespoons': 16,
            'teaspoon': 48, 'teaspoons': 48,
            'ounce': 8, 'ounces': 8, 'pound': .5, 'pounds': .5}

baseline_tagger = nltk.UnigramTagger(model=allDict)

ingrList = []

with open('taggedIngredients.csv','w',newline='') as tagFile:
  writer=csv.writer(tagFile)

  with open('fixedIngredients.csv', 'r') as ingrFile:
    ingr_File_text = csv.reader(ingrFile,delimiter = ',')

    for line in ingr_File_text:
      recLine = []
      for ingr in line:
        ingr = ingr.replace(",", "")
        ingr = ingr.replace('\xa0', ' ')
        ingr = re.sub(r'\(.*\)','',ingr)
        newLine = baseline_tagger.tag(ingr.split(' '))     

        for i in range(len(newLine)):
          if newLine[i][1] is None:
            newLine[i] = (newLine[i][0],'NONE')

        ingrList.append(newLine)

        newLineCopy = []
        dbl = False
        temp_tag = None
        for tag_ing in newLine:

          if tag_ing[1] != 'CD':
            if temp_tag:
              try:
                newLineCopy.append((str(round(float(eval(temp_tag[0])),3)),'CD'))
              except:
                newLineCopy.append((card_str_to_int[temp_tag[0]],'CD'))
            dbl = False
            temp_tag = None
            newLineCopy.append(tag_ing)
        
          else: #tag is 'CD'
            if not dbl:
              dbl = True
              temp_tag = tag_ing
            else: #dbl is true => second cardinal
              total = str(round(float(temp_tag[0])+float(eval(tag_ing[0])),3))
              newLineCopy.append((total,'CD'))
              dbl = False
              temp_tag = None



        newLine = [] 
        temp_tag = None
        combflag = False
        for tag_ing in newLineCopy:

          if tag_ing[1] != 'CD':
            if temp_tag:
              if tag_ing[0] == 'to':
                combflag = True #don't store the 'to'
              else:
                combflag = False
                newLine.append(temp_tag)
                temp_tag = None
                newLine.append(tag_ing)
            else:
              newLine.append(tag_ing)
        
          else: #tag is 'CD'
            if not combflag:
              temp_tag = tag_ing
            else: #combine
              avg = str(round((float(temp_tag[0])+float(tag_ing[0]))/2,3))
              newLine.append((avg,'CD'))
              combflag = False
              temp_tag = None

        
        newLineCopy = [] 
        addflag = breakflag = skipflag = False
        contLine = []
        
        for tag_ing in newLine:
          #print('before',tag_ing)
          if skipflag:
            if 'round' in tag_ing[0]:
              continue
            skipflag = False
            w /= cup_conv_p[tag_ing[0]]
            newLineCopy.append((str(round(v+w,3)),'CD'))
            newLineCopy.append(('cups', 'UNT'))
            continue

          if breakflag:
            if tag_ing[0] in ['additional','more','extra']:
              break
            else:
              contLine.append(newLineCopy)
              newLineCopy = []
              breakflag = False
            
          if addflag:
            w = eval(tag_ing[0])
            addflag = False
            skipflag = True
            continue

          if tag_ing[0] == 'plus':
            if not newLineCopy or newLineCopy[-1][1] != 'UNT':
              breakflag = True
            else:
              addflag = True
              u = newLineCopy.pop()[0]
              v = eval(newLineCopy.pop()[0])/cup_conv_p[u]
            continue
          
          if tag_ing[1] != 'NONE':
            newLineCopy.append(tag_ing)
          
        contLine.append(newLineCopy)
        
        for newLineCopy in contLine:
          ingCount = 0
          firstIngPos = None
          vanflag = False
          lemflag = False
          for i in range(len(newLineCopy)):
            if 'lemon' in newLineCopy[i][0]:
              lemflag = True
              #print('l',newLineCopy)
            if 'vanilla' in newLineCopy[i][0]:
              vanflag = True
              #print('v',newLineCopy)
            if newLineCopy[i][1] == 'ING':
              ingCount += 1
              if ingCount == 1:
                firstIngPos = i
            if newLineCopy[i][1] == 'UNT' and newLineCopy[i][0] in cup_conv:
              u = newLineCopy[i][0]
              j = i-1
              if newLineCopy[i-1][0] in ['heaping','scant']:
                j -= 1
              if newLineCopy[j][1] == 'CD':
                newLineCopy[j] = (str(round(eval(newLineCopy[j][0])/cup_conv[u],3)), 'CD')
              newLineCopy[i] = ('cups', 'UNT')

          if ingCount >= 2: 
            if vanflag:
              newLineCopy = newLineCopy[-3:]
              ingCount = 1
            elif lemflag:
              ingCount = 1
            else:
              del newLineCopy[firstIngPos]
        if ingCount > 3: 
          break

        for newLineCopy in contLine:
          if len(newLineCopy) > 1:     
            recLine.append(newLineCopy)
        #end new stuff

      #print(recLine)
      writer.writerow(recLine)


with open('ingrText.txt','w') as ingrText:

  with open('taggedIngredients.csv', 'r') as taggedIngr:
    taggedIngr_file = csv.reader(taggedIngr,delimiter = ',')

    for recipe in taggedIngr_file:
      ingrText.write("Ingredients:")
      ingrText.write("\n")


      for line in recipe:
        line = ast.literal_eval(line)
        for item in line: 
          ingrText.write(item[0].replace('_'," "))
          ingrText.write(' ')
        ingrText.write('\n')
      ingrText.write('\n')
      ingrText.write('\n')
```

Set up our RNN

```
text = open('ingrText.txt', 'rb').read().decode(encoding='utf-8')

vocab = sorted(set(text))
#print ('{} unique characters'.format(len(vocab)))

char2idx = {u:i for i, u in enumerate(vocab)}
idx2char = np.array(vocab)

text_as_int = np.array([char2idx[c] for c in text])

'''
#Visualize mapping:
print('{')
for char,_ in zip(char2idx, range(20)):
    print('  {:4s}: {:3d},'.format(repr(char), char2idx[char]))
print('  ...\n}')

print ('{} ---- characters mapped to int ---- > {}'.format(repr(text[:13]), text_as_int[:13]))
'''

# The maximum length sentence we want for a single input in characters
seq_length = 100
examples_per_epoch = len(text)//(seq_length+1)

# Create training examples / targets
char_dataset = tf.data.Dataset.from_tensor_slices(text_as_int)

sequences = char_dataset.batch(seq_length+1, drop_remainder=True)

'''
#More visualizing:
sequences = char_dataset.batch(seq_length+1, drop_remainder=True)

for item in sequences.take(5):
  print(repr(''.join(idx2char[item.numpy()])))
'''

def split_input_target(chunk):
  input_text = chunk[:-1]
  target_text = chunk[1:]
  return input_text, target_text

dataset = sequences.map(split_input_target)

BATCH_SIZE = 64

BUFFER_SIZE = 10000

dataset = dataset.shuffle(BUFFER_SIZE).batch(BATCH_SIZE, drop_remainder=True)

vocab_size = len(vocab)

embedding_dim = 256

rnn_units = 1024

def build_model(vocab_size, embedding_dim, rnn_units, batch_size):
  model = tf.keras.Sequential([
    tf.keras.layers.Embedding(vocab_size, embedding_dim,
                              batch_input_shape=[batch_size, None]),
    tf.keras.layers.GRU(rnn_units,
                        return_sequences=True,
                        stateful=True,
                        recurrent_initializer='glorot_uniform'),
    tf.keras.layers.Dense(vocab_size)
  ])
  return model

model = build_model(
    vocab_size = len(vocab),
    embedding_dim=embedding_dim,
    rnn_units=rnn_units,
    batch_size=BATCH_SIZE)
```

```
for input_example_batch, target_example_batch in dataset.take(1):
  example_batch_predictions = model(input_example_batch)
  print(example_batch_predictions.shape, "# (batch_size, sequence_length, vocab_size)")
```

```
def loss(labels, logits):
  return tf.keras.losses.sparse_categorical_crossentropy(labels, logits, from_logits=True)

example_batch_loss  = loss(target_example_batch, example_batch_predictions)
print("Prediction shape: ", example_batch_predictions.shape, " # (batch_size, sequence_length, vocab_size)")
print("scalar_loss:      ", example_batch_loss.numpy().mean())
```

```
model.compile(optimizer='adam', loss=loss)
```

```
# Directory where the checkpoints will be saved
checkpoint_dir = './training_checkpoints'
# Name of the checkpoint files
checkpoint_prefix = os.path.join(checkpoint_dir, "ckpt_{epoch}")

checkpoint_callback=tf.keras.callbacks.ModelCheckpoint(
    filepath=checkpoint_prefix,
    save_weights_only=True)
```

```
def generate_text(model, start_string):

  num_generate = 2000

  input_eval = [char2idx[s] for s in start_string]
  input_eval = tf.expand_dims(input_eval, 0)

  text_generated = []

  temperature = 1.0

  # Here batch size == 1
  model.reset_states()
  for i in range(num_generate):
    predictions = model(input_eval)
    # remove the batch dimension
    predictions = tf.squeeze(predictions, 0)

    # using a categorical distribution to predict the character returned by the model
    predictions = predictions / temperature
    predicted_id = tf.random.categorical(predictions, num_samples=1)[-1,0].numpy()

    # We pass the predicted character as the next input to the model
    # along with the previous hidden state
    input_eval = tf.expand_dims([predicted_id], 0)

    text_generated.append(idx2char[predicted_id])

  return (start_string + ''.join(text_generated))
```
